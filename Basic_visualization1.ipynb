```python
youtube_data = pd.read_csv('/content/drive/MyDrive/Project-EMSE6574/youtube_data.csv')

youtube_data = youtube_data.dropna()
```

```python
youtube_data['trending_date'] = pd.to_datetime(youtube_data['trending_date'], format = "%y.%d.%m")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
filtered_df = youtube_data.loc[(youtube_data['trending_date'] == '2017-11-14')]
```

```python
filtered_df['log_views'] = np.log10(youtube_data['views'])
filtered_df['log_likes'] = np.log10(youtube_data['likes'])
filtered_df['log_dislikes'] = np.log10(youtube_data['dislikes'])
filtered_df['log_comment'] = np.log10(youtube_data['comment_count'])
```
